```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
atlantic_data = pd.read_parquet('../data/atlantic_flyway_corridor.parquet')
atl_data = atlantic_data[
    (atlantic_data['migration_end_raw'].astype(str).str.contains('Oct 28'))
]
atl_data = atl_data[
    (atl_data['region_name'].str.contains('Duval')) | (atl_data['region_name'].str.contains('Essex County, New Jersey'))
]
atl_data
```

```python
pacific_data = pd.read_parquet('../data/pacific_flyway_corridor.parquet')
pac_data = pacific_data[
    (pacific_data['migration_end_raw'].astype(str).str.contains('Oct 28'))
]
pac_data = pac_data[
    (pac_data['region_name'].str.contains('Contra Costa'))
]
pac_data
```

```python
update_data = pd.concat([atl_data, pac_data])
update_data
```

```python
birdcast_data = pd.read_parquet('../data/birdcast_data.parquet')
birdcast_data = birdcast_data.dropna(subset=['total_birds']).reset_index(drop=True)
birdcast_data['migration_start_utc'] = pd.to_datetime(birdcast_data['migration_start_utc'])
birdcast_data['migration_end_utc'] = pd.to_datetime(birdcast_data['migration_end_utc'])
birdcast_data.sort_values(by='migration_start_utc', inplace=True)
birdcast_data
```

```python
all_data = pd.concat([birdcast_data, update_data])
all_data
```

```python
for col in all_data.columns:
    if ("utc" in col) or ("scrape" in col):
        # all_data[col] = [val[:val.find("-")+6] if "-" in val else val for val in all_data[col]]
        # all_data[col] = pd.to_datetime(all_data[col])
        all_data[col] = all_data[col].astype(str)
# all_data.to_parquet('../data/birdcast_data.parquet', compression='snappy', index=False)
all_data.to_csv('../data/birdcast_data.csv', index=False)
all_data
```
